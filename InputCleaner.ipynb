## Imports

```python
import pandas as pd
```

## Data import

```python
base_data = pd.read_fwf("./Data/Calibration.txt")
first_data = pd.read_fwf("./Data/Run1.txt")
second_data = pd.read_fwf("./Data/Run2.txt")
```

## Line Spliting 

```python
base_data = base_data.iloc[5:]
first_data = first_data.iloc[5:]
second_data = second_data.iloc[5:]
```

```python
base_data
```

```python
base_data.columns = ["Data"]
first_data.columns = ["Data"]
second_data.columns = ["Data"]
```

```python
base_data = base_data.Data.str.split(" ",expand=True,)
first_data = first_data.Data.str.split(" ",expand=True,)
second_data = second_data.Data.str.split(" ",expand=True,)
```

## Deleting the first column

```python
del base_data[base_data.columns[0]]
del first_data[first_data.columns[0]]
del second_data[second_data.columns[0]]
```
